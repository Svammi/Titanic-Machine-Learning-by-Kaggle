метод **info()** показывает информацию о данных: название столбцов, колличество не Null элементов в каждом столбце и тип Dtype столбца

```python
print(test_data.describe()) #описательная статистика
```

Метод **describe()** выдает описательную статистику по колонкам. При этом учитываются только числовые типы данных. Выводи следующие статистические данные: 

- count - количество значений в столбце (не Null)
- mean - среднее значение для данной колонки, вычисляется как среднее арифметическое
- std - стандартное отклонение. Показывает как каждое значение столбца отклоняется от среднего (mean). [Подробнее](https://exceltip.ru/%D1%87%D1%82%D0%BE-%D1%82%D0%B0%D0%BA%D0%BE%D0%B5-%D1%81%D1%82%D0%B0%D0%BD%D0%B4%D0%B0%D1%80%D1%82%D0%BD%D0%BE%D0%B5-%D0%BE%D1%82%D0%BA%D0%BB%D0%BE%D0%BD%D0%B5%D0%BD%D0%B8%D0%B5-%D0%B8%D1%81%D0%BF/)
- min - минимальное значение в столбце
- 25% - нижний процентиль или квантиль. 
- 50% - средний процентиль, совпадает с mean
- 75% - верхний процентиль.
- max - максимальное значение в столбце

```python
# количество неопределнных значений
print(test_data.isnull().sum())
```

Метод **isnull()** определяет False/True для Null значений

Подсчитаем сколько в столбцах есть неопределленных значений - Null

Необходимо применить метод **экстраполяции** для определения значения Null-вых ячеек.

```python
alives = test_data[test_data['Survived'] == 1] # список выживших пассажиров
died = test_data[test_data['Survived'] == 0] # список умерших пассажиров
print(alives["Age"])
```

```python
plt.subplot(1,2,1, title = "Alive")
plt.hist(alives["Age"], bins=range(0,86,5))
plt.subplot(1,2,2, title = "Died")
plt.hist(died["Age"],bins=range(0,86,5))
plt.show()
```

```python
# найдем долю выживаемости и смертности среди детей, молодых и взрослых
children = test_data.loc[test_data["Age"] <= 15]
young = test_data.loc[(15 < test_data["Age"]) & ( test_data["Age"] <= 40 )]
adult = test_data.loc[40 < test_data["Age"]]

def rate_survival(data):
    died, alive = len(data[data["Survived"] == 0]) / len(data), len(data[data["Survived"] == 1]) / len(data)
    return [died, alive]

rate_survival_shildren = rate_survival(children)
rate_survival_young = rate_survival(young)
rate_survival_adult = rate_survival(adult)

rate_died_alive_age = np.array([rate_survival_shildren, rate_survival_young, rate_survival_adult]).transpose()
print(rate_died_alive_age)
```

```python
labels = ["Children (0-15)", "Young(16-40)", "Adult(older 40)"]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, rate_died_alive_age[0], width, label='Died')
rects2 = ax.bar(x + width/2, rate_died_alive_age[1], width, label='Alive')
ax.set_xticks(x)
ax.set_xticklabels(labels)
fig.tight_layout()

plt.show()
```

Из гистограммы видно, что самая высока выживаемость среди детей до 15 лет. Самый низкий показатель выживаемости у взрослых старше 40 лет. Однако есть 80-ти летний человек, которому удалось выжить


```python
# диаграмма рассеяния
import seaborn as sb
sb.swarmplot(x = "Survived", y = "Age", data = test_data)
plt.show()
```

```python
# проведем анализ смертности по половому признаку
print(test_data["Sex"].unique())
male = test_data[test_data["Sex"] == 'male']
female = test_data[test_data["Sex"] == 'female']

rate_male = rate_survival(male)
rate_female = rate_survival(female)

rate_died_alive_sex = np.array([rate_male, rate_female]).transpose()
print(rate_male, rate_female)
```

```python
labels = ["Male", "Female"]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, rate_died_alive_sex[0], width, label='Died')
rects2 = ax.bar(x + width/2, rate_died_alive_sex[1], width, label='Alive')
ax.set_xticks(x)
ax.set_xticklabels(labels)
fig.tight_layout()

plt.show()
```

Из гистограммы видно, что большая смертность среди мужчин и высокая выживаемость среди женщин. 

```python
# таже самая диаграмма, но по-питоновски с применением библиотеки SEABORN
g = sb.catplot(x="Survived", col="Sex", col_wrap=4, data=test_data, kind="count")
plt.show()
```

```python
# рассмотрим как Класс билета влияет на смертность
print(test_data["Pclass"].unique())
sb.catplot(x="Survived", col="Pclass", col_wrap=3, data=test_data, kind="count")
plt.show()
```

Из гистограмм видно, что высокая смертность встречается среди пассажиров 3-го класса, и в тоже время наблюдаем высокую выживаемость среди пассажиров 1-го класса.
